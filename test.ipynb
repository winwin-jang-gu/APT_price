```python
import pandas as pd
```

##### 월평균 임금

```python
df1 = pd.read_excel('./월평균_임금.xlsx')
df1.head(3)
```

```python
df1 = df1.melt(id_vars=['행정구역별(1)','항목'])
df1
```

```python
df1 = pd.to_numeric(df1['value'], errors='coerce')
df1
```

```python
df1.isna().sum()
```
